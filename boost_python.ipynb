```python
from tls_crypt import *

def str2list(s) :
    r = []
    for c in s: r.append(ord(c))
    return r

def add_padding(l):
    padding = 15 - len(l) % 16
    for i in range(padding + 1): l.append(padding)
    return l

def remove_padding(l):
    padding = l[-1] + 1
    for i in range(padding): l.pop()
    return l

def list2str(l) :
    r = ''
    for c in l: r += chr(c)
    return r

def hex2list(s):
    assert len(s) % 16 == 0
    l = []
    str = ''
    for c in s:
        str += c
        if len(str) == 2:
            l.append(int(str, 16))
            str = ''
    return l
```
